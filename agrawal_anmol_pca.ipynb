```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Perform PCA with 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# a) Plot 2D scatter plot
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title('2D Scatter Plot of Iris Dataset with PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(*scatter.legend_elements(), title='Classes')
plt.show()

# b) Calculate total percentage of variance captured by the first 2 components
explained_variance_ratio = pca.explained_variance_ratio_
total_variance_captured = np.sum(explained_variance_ratio)
print('Total Percentage of Variance Captured by the First 2 Components: {}%'.format(total_variance_captured * 100))

# c) Display the strength of each of the two principal components
component_strengths = pca.singular_values_
print('Strength of Principal Component 1: {}'.format(component_strengths[0]))
print('Strength of Principal Component 2: {}'.format(component_strengths[1]))

# d) Display the magnitude of each of the two principal components
component_magnitude_1 = np.linalg.norm(X_pca[:,0])
component_magnitude_2 = np.linalg.norm(X_pca[:, 1])
print('Magnitude of Principal Component 1: {}'.format(component_magnitude_1))
print('Magnitude of Principal Component 2: {}'.format(component_magnitude_2))
```
